```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Use LaTEX font
from matplotlib import rc
rc('font', **{'family': 'serif', 'serif': ['Palatino'], 'size': 20})
```

```python
df = pd.read_csv("svmdoc2vec_results.csv",header=None)
df = df[[0,1,2]]
df = df.sort_values(by=[1,0])
df_dm_dbow = df[df[0] == "concat_dm_dbow"]
df_dm = df[df[0] == "dm"]
df_dbow = df[df[0] == "dbow"]
df_w_dbow = df[df[0] == "w_dbow"]
```

```python
def plot_df(df, fig, ax):
    x = np.log2(df[[1]] )
    y = df[[2]]
    ax.plot(x, y, marker='.')
    ax.set_ylabel("Accuracy")
    ax.set_xlabel("Size of embedding vector") # log$_2$|$D$|
    ax.set_xticks(range(0,11))
    ax.set_xticklabels(df[[1]].values.flatten())

    # secax = ax.secondary_xaxis('top', functions=(lambda x: x, lambda x: x))
    # secax.set_xlabel("Embedding dimension-size, $D$")
    # secax.set_xticks(range(0,11))
    # secax.set_xticklabels(df[[1]].values.flatten())
    
```

```python
fig, ax = plt.subplots(constrained_layout=False)
fig.set_figwidth(8)
fig.set_figheight(6)
plot_df(df_w_dbow, fig, ax)
plot_df(df_dm, fig, ax)
# plot_df(df_dm_dbow, fig, ax)

plt.ylim(0.4,0.9)
plt.grid(True, alpha=0.2)

plt.legend(["PV-DBOW", "PV-DM"])#, "DM+DBOW"])
# plt.title("Accuracy against embedding size for DBOW and DM doc2vec variants")
```
